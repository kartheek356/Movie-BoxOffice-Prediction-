```python
from sklearn.impute import KNNImputer

# Selecting numerical features for estimating budget and revenue
numeric_features = ["popularity", "vote_average", "vote_count"]

# Convert genre into a numerical feature by counting the number of genres per movie
df["genre_count"] = df["genre_names"].apply(lambda x: len(eval(x)) if isinstance(x, str) else 0)

# Convert text-based 'overview' into a numerical feature based on length
df["overview_length"] = df["overview"].apply(lambda x: len(str(x)))

# Features to be used for imputing budget and revenue
features_for_imputation = ["popularity", "vote_average", "vote_count"]

# Impute missing values using KNN
imputer = KNNImputer(n_neighbors=5)
imputed_values = imputer.fit_transform(df[features_for_imputation])

# Normalize budget and revenue based on estimated values
df["budget"] = imputed_values[:, 0] * 100000  # Scaling based on trend
df["revenue"] = imputed_values[:, 0] * 150000  # Scaling based on trend

df.head()
```

# Outliers


```python
import numpy as np

# Identify numerical columns
numerical_columns = df.select_dtypes(include=[np.number]).columns

# Detect outliers using the IQR (Interquartile Range) method
outlier_dict = {}

for col in numerical_columns:
    Q1 = df[col].quantile(0.25)  # First quartile (25th percentile)
    Q3 = df[col].quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1  # Interquartile range

    # Define outlier bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Count the number of outliers
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_dict[col] = len(outliers)

# Display the number of outliers per column
outlier_dict
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Identify numerical columns
numerical_columns = df.select_dtypes(include=["number"]).columns

# Set up subplots
num_cols = len(numerical_columns)
fig, axes = plt.subplots(nrows=(num_cols // 3) + 1, ncols=min(3, num_cols), figsize=(15, 5 * ((num_cols // 3) + 1)))

# Flatten axes for easy iteration
axes = axes.flatten() if num_cols > 1 else [axes]

# Plot a boxplot for each numerical column
for i, col in enumerate(numerical_columns):
    sns.boxplot(x=df[col], ax=axes[i])
    axes[i].set_title(f"Boxplot of {col}")
    axes[i].set_xlabel(col)

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

This visualization consists of boxplots, each representing the distribution of different numerical features in the dataset. Boxplots help in identifying outliers, data spread, and skewness in variables like popularity, vote count, budget, and revenue.
